```python
import cv2
from ultralytics import YOLO
from IPython.display import display, Image
import os
import time
```

### Leer camaras

```python
def find_cameras():
    index = 0
    arr = []
    while True:
        cap = cv2.VideoCapture(index)
        if not cap.read()[0]:
            break
        else:
            arr.append(index)
        cap.release()
        index += 1
    return arr

cameras = find_cameras()
print("Cámaras disponibles:", cameras)
```

### Realizar prediccion

```python
model = YOLO("prototype_2.pt")

cam_index = 1  # Cambia este índice según tu necesidad

cap = cv2.VideoCapture(cam_index)

output_folder = "fotos_tomadas"
predicted_folder = "fotos_predecidas"

os.makedirs(output_folder, exist_ok=True)
os.makedirs(predicted_folder, exist_ok=True)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    if frame_count % 1 == 0:  
        image_path = os.path.join(output_folder, f"frame_{frame_count}.jpg")
        cv2.imwrite(image_path, frame)
        
        results = model.predict(image_path)

        detected_classes = [] 
        names = {0: 'arma_detectada'} 

        for result in results:
            boxes = result.boxes
            for box in boxes:
                class_id = int(box.cls)  
                class_name = names.get(class_id, "Unknown")  
                detected_classes.append(class_name)  
                
                x1, y1, x2, y2 = box.xyxy[0]
                confidence = box.conf[0]
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                cv2.putText(frame, f"{class_name}: {confidence:.2f}", (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        
        predicted_image_path = os.path.join(predicted_folder, f"{frame_count}.jpg")
        cv2.imwrite(predicted_image_path, frame)

        cv2.imshow('Selected Camera', frame)
        
        #print(detected_classes)
    
    frame_count += 1

    if cv2.getWindowProperty('Selected Camera', cv2.WND_PROP_VISIBLE) < 1:
        break
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
