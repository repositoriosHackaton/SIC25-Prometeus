### detectar videos guardados en la carpeta video_dataset, habian varios videos

```python
import os

# Especificar la ruta de la carpeta
carpeta = 'video'

# Obtener la lista de archivos en la carpeta
archivos = os.listdir(carpeta)

# Filtrar solo los archivos (excluye subcarpetas)
solo_archivos = [archivo for archivo in archivos if os.path.isfile(os.path.join(carpeta, archivo))]

# Contar el número de archivos
numero_archivos = len(solo_archivos)

print(f"Hay {numero_archivos} archivos en la carpeta '{carpeta}'")
print(solo_archivos)
```

### Desfragmentar video en varios frames

```python
import cv2
import os

for i in range(numero_archivos):
    video = 1+i
    # Ruta del video
    video_path = f"video/video ({video}).mp4"
    # Carpeta donde se guardarán las imágenes
    output_folder = f"output_images_{video}"
    
    # Crear la carpeta si no existe
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Cargar el video
    cap = cv2.VideoCapture(video_path)
    
    # Verificar si el video se ha cargado correctamente
    if not cap.isOpened():
        print("Error: No se pudo abrir el archivo de video.")
    else:
        frame_count = 0
        while True:
            ret, frame = cap.read()
            if not ret:
                break
    
            # Guardar cada cuadro como una imagen
            frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.jpg")
            cv2.imwrite(frame_filename, frame)
            frame_count += 1
    
        cap.release()
        print(f"Se han guardado {frame_count} cuadros en la carpeta '{output_folder}'")
    
```

### En esta parte, hay dos listas adonde, en una esta la lista de archivos completa, en la otra esta los que deseo tener, pero como hay varias carpetas y solamente quiero que me elimine los archivos individualmente por cada carpeta de video tal...

```python
import os

# Especificar la ruta de la carpeta
carpeta = 'data/'

# Obtener la lista de archivos en la carpeta
archivos = os.listdir(carpeta)
#print(archivos)

# Filtrar solo los archivos (excluye subcarpetas)
#archivos = [archivo for archivo in archivos if os.path.isfile(os.path.join(carpeta, archivo))]

diccionario_completo = {}
diccionario_archivos_necesarios = {}
for i in archivos:
    #print(i)
    lista = os.listdir(f"{carpeta}/{i}")
    diccionario_completo[i] = lista
    diccionario_archivos_necesarios[i] = lista[::5]
    
    #print(lista[::5])

# Crear un nuevo diccionario para almacenar los resultados
resultado_diccionario = {}

# Iterar sobre las claves de diccionario1
for clave in diccionario_completo:
    # Obtener la lista correspondiente de cada diccionario
    lista1 = diccionario_completo[clave]
    lista2 = diccionario_archivos_necesarios.get(clave, [])  # Obtener la lista de diccionario2, o una lista vacía si no existe la clave
    
    # Crear una nueva lista eliminando los elementos de lista1 que están en lista2
    nueva_lista = [item for item in lista1 if item not in lista2]
    
    # Almacenar la nueva lista en el resultado_diccionario
    resultado_diccionario[clave] = nueva_lista

print(resultado_diccionario)

```

### Eliminar objetos de la lista pasada

```python
import os

# Ruta base de tus carpetas
ruta_base = "data"

# Iterar sobre cada carpeta y sus archivos a eliminar
for carpeta, archivos in resultado_diccionario.items():
    for archivo in archivos:
        ruta_completa = os.path.join(ruta_base, carpeta, archivo)
        if os.path.exists(ruta_completa):
            os.remove(ruta_completa)
            print(f"El archivo {ruta_completa} ha sido eliminado.")
        else:
            print(f"El archivo {ruta_completa} no existe.")
```
